#### Problema No. 38 ($Pescador$).

```python
# corrida 1: Captura de la Trucha y Salida Voluntaria
# --- SIMULADOR DE IA: EL PESCADOR Y EL PEZ ---

# 1. Configuración inicial de la simulación
print("--- BIENVENIDO AL SIMULADOR DE PESCA INTELIGENTE ---")
energia = 100         # El pescador tiene energía limitada
cebo = 1              # 1: Gusano (Básico), 2: Señuelo (Avanzado)
profundidad = 5       # Profundidad inicial en metros
jugando = True        # Variable de control para el ciclo while

# 2. Ciclo principal interactivo (Menú)
while jugando == True and energia > 0: # El ciclo se repite mientras haya energía [3, 6]
    print(f"\nEstado: Energía {energia} | Cebo: {cebo} | Profundidad: {profundidad}m")
    print("MENÚ DE ACCIONES:")
    print("1. Cambiar cebo (Gusano <-> Señuelo)")
    print("2. Ajustar profundidad (Subir/Bajar)")
    print("3. Lanzar el anzuelo (Interacción con el Pez)")
    print("4. Salir de la simulación")

    opcion = int(input("Selecciona una opción (1-4): ")) # Entrada del usuario [7, 8]

    # --- Lógica de las opciones ---
    if opcion == 1:
        if cebo == 1:
            cebo = 2
            print("IA Mensaje: Has equipado el Señuelo Brillante.")
        else:
            cebo = 1
            print("IA Mensaje: Has vuelto a usar Gusanos orgánicos.")
        energia = energia - 5 # Operador aritmético: cada acción gasta energía [9, 10]

    elif opcion == 2:
        nueva_prof = int(input("Ingresa profundidad deseada (1-20 metros): "))
        if nueva_prof >= 1 and nueva_prof <= 20: # Operador relacional y lógico [11, 12]
            profundidad = nueva_prof
            print(f"IA Mensaje: Anzuelo posicionado a {profundidad}m.")
        else:
            print("IA Error: Esa profundidad está fuera de rango.")
        energia = energia - 5

    elif opcion == 3:
        print("Lanzando anzuelo... La IA del Pez está evaluando...")

        # --- Lógica de la Inteligencia Artificial del Pez ---
        # El pez analiza el tipo de cebo y la profundidad para decidir su comportamiento [13, 14]
        if cebo == 1 and profundidad < 10:
            print("IA del Pez: ¡Mmm, un gusano cerca de la superficie! ¡PICÓ!")
            print("Resultado: Has capturado una Trucha pequeña.")
        elif cebo == 2 and profundidad >= 10:
            print("IA del Pez: Detecto algo brillante en las profundidades... ¡ES UNA PRESA!")
            print("Resultado: ¡HAS CAPTURADO UN GRAN PEZ DORADO!")
        elif cebo == 1 and profundidad >= 10:
            print("IA del Pez: Un gusano aquí abajo se ve muy sospechoso. Lo ignoro.")
            print("Resultado: El pez no mordió el anzuelo.")
        else:
            print("IA del Pez: No hay nada interesante en esta zona.")
            print("Resultado: Sigue intentándolo.")

        energia = energia - 20 # Lanzar el anzuelo gasta más energía

    elif opcion == 4:
        print("IA Mensaje: Finalizando simulación. ¡Gracias por pescar!")
        jugando = False # Condición de salida para terminar el ciclo while [15, 16]

    else:
        print("IA Error: Opción no válida en el menú.")

# 3. Análisis final de la simulación
if energia <= 0:
    print("\n--- ¡TE HAS QUEDADO SIN ENERGÍA! ---")
    print("La simulación termina porque el pescador está cansado.")

print("--- FIN DE LA SESIÓN ---")
print("Programa realizado por Pedro Martinez NL = 38 =)")
```
